##### 그래프의 종류
**막대 그래프** : 데이터를 카테고리별로 값을 막대로 표현하고자 할 때 사용하는 그래프 형식 (수직 막대, 수평 막대 - 그래프)
> 카테고리별 데이터를 비교하고자 할 때 사용 (제품 판매량, 지역별 인구수)
**선 그래프** : 시간의 흐음에 따른 데이터 포인터를 선으로 연결하여 사용하는 그래프 형식
>하나의 요소에 대해 시간의 흐름에 따라 변경값을 나타내고자 할때 사용 (기온, 주식)
**파이 차트** : 전체 데이터에 대한 각 요소의 비율을 원형으로 나타낼때 사용하는 그래프
>전체에 대한 요소의 비율을 표현하고자 할 때 사용 (투표결과 , 전체가 고정이고 비율을 볼때)
**산점도(점 그래프)** : 두 변수 간의 상관관계를 표현할때 사용하는 그래프 형식
>각 요소간 관계성을 파악할때 사용 (키와 몸무게 비교)
**히스토그램** : 범위별로 그룹화하여 각 그룹에 속하는 데이터의 수를 막대로 표현하는 그래프 형식
> 데이터의 분포를 비교할때 사용 (시험점수분포)
**영역 그래프** : 선 그래프의 아래쪽 빈 공간을 채운 그래프
> 시간의 흐름에 따라 데이터를 비교할 때 사용 , 변화량을 강조하고자 할때 사용
**박스 그래프** : 데이터의 분포를 나타낼때 사용하는 그래프 형식
> 최대값, 최소값, 사분위수로 데이터의 이상치를 판단할때

````python
plt.plot([1,2,4,6,16])
plt.show()
````

````python
#그래프의 제목 지정 title()메서드 사용
#plot(x축에 대한 리스트, y축에 대한 리스트)
plt. title('x 추가')
plt.plot([2020,2021,2022,2023,2024],[1,2,4,8,16])
plt.show
````

````python
# 차트 스타일 지정
# plot() 메서드의 세번째 매개변수로 지정된 문자열 포멧으로 전달하여 스타일 지정 가능
# color (c) : 선 색상, linewidth (lw) : 선 굵기 , linestyle (ls) : 선 스타일
# marker : 마커 종류 , markersize(ms): 마커크기, markeredgecolor(mec): 마커테두리 색상
#markeredgewidth(mew) : 마커 테두리 굴기 markerfacecolor(mfc): 마커 내부 색상
plt.title('c2-- 스타일 적용')
plt.plot([2020,2021,2022,2023,2024],[1,2,4,8,16],'c2--')
plt.show
````

````python
# 차트 스타일 지정
# plot() 메서드의 세번째 매개변수로 지정된 문자열 포멧으로 전달하여 스타일 지정 가능
# color (c) : 선 색상, linewidth (lw) : 선 굵기 , linestyle (ls) : 선 스타일
# marker : 마커 종류 , markersize(ms): 마커크기, markeredgecolor(mec): 마커테두리 색상
#markeredgewidth(mew) : 마커 테두리 굴기 markerfacecolor(mfc): 마커 내부 색상

plt.title('스타일 적용')
plt.plot([2020,2021,2022,2023,2024],[1,2,4,8,16],c='#00998850', lw = 5, ls = ':', marker='*',
         ms=10, mfc = '#009988')
plt.show
````

````python
# 차트 범위 지정
#xlim(), ylin() 메서드로 x축과 y축의 그래프의 범위를 지정할 수 있다 

plt.title('스타일 적용')
plt.plot([2020,2021,2022,2023,2024],[1,2,4,8,16],c='#00998850', lw = 5, ls = ':', marker='*',
         ms=10, mfc = '#009988')
plt.xlim(2019, 2025)
plt.ylim(3, 20)
plt.show
````

````python
# 틱 지정
# xticks(), yticks()메서드로 틱의 위치나 라벨 지정 가능
plt.title('틱 적용')
plt.plot([2020,2021,2022,2023,2024],[1,2,4,8,16],c='#00998850', lw = 5, ls = ':', marker='*',
         ms=10, mfc = '#009988')
plt.xticks([2020,2022,2024])
plt.yticks([0,8,16],['저점','중간점','고점'])
plt.show
````

````python
#그리드
#grid() 메서드로 그리드를 표시할지 경절할 수 있음

plt.title('그리드 적용')
plt.plot([2020,2021,2022,2023,2024],[1,2,4,8,16],c='#00998850', lw = 5, ls = ':', marker='*',
         ms=10, mfc = '#009988')
plt.xticks([2020,2022,2024])
plt.yticks([0,8,16],['저점','중간점','고점'])
plt.grid(True)
plt.show
````

````python
#차트 겹치기
#plot()메서드를 중복해서 호출하여 차트를 겹칠수 있다
plt.title('차트 겹치기')
plt.plot([2020,2021,2022,2023,2024],[1,2,4,8,16],c='#00998850', lw = 5, ls = ':', marker='*',
         ms=10, mfc = '#009988')
plt.plot([2020,2021,2022,2023,2024],[16,8,4,2,1],c='#88990050', lw = 5, ls = ':', marker='*',
         ms=10, mfc = '#889900')
plt.xticks([2020,2022,2024])
plt.yticks([0,8,16],['저점','중간점','고점'])
plt.grid(True)
plt.show
````

````python
#범례 표시
#egend()메서드를 통해 범례를 표시할수 있음
#loc매개변수에 위치에 해당하는 정수를 전달하여 위치 지정 가능
plt.title('범례 표시')
plt.plot([2020,2021,2022,2023,2024],[1,2,4,8,16],c='#00998850', lw = 5, ls = ':', marker='*',
         ms=10, mfc = '#009988', label = '증가')
plt.plot([2020,2021,2022,2023,2024],[16,8,4,2,1],c='#88990050', lw = 5, ls = ':', marker='*',
         ms=10, mfc = '#889900', label = '감소')
plt.xticks([2020,2022,2024])
plt.yticks([0,8,16],['저점','중간점','고점'])
plt.grid(True)
plt.legend(loc=0)
plt.show
````

````python
#각 틱 라벨
#xlabel(), ylabel() 메서드로 x축, y축의 라벨을 표시 가능
plt.title('각 틱 라벨')
plt.plot([2020,2021,2022,2023,2024],[1,2,4,8,16],c='#00998850', lw = 5, ls = ':', marker='*',
         ms=10, mfc = '#009988', label = '증가')
plt.plot([2020,2021,2022,2023,2024],[16,8,4,2,1],c='#88990050', lw = 5, ls = ':', marker='*',
         ms=10, mfc = '#889900', label = '감소')
plt.xticks([2020,2022,2024])
plt.yticks([0,8,16],['저점','중간점','고점'])
plt.xlabel('연도')
plt.ylabel('가치')
plt.grid(True)
plt.legend(loc=0)
plt.show
````

### 막대 그래프
x축의 값이 분류가 가능한 카테고리 형태일때 bar(), barh() 메서드로 막대 그래프를 표현할 수 있음

````python
plt.title("막대 그래프")
plt.bar(['a','b','c'],[20,78,43])
plt.show
````

````python
plt.title("막대 그래프")
plt.bar(['a','b','c'],[20,78,43],color = '#ff0000')
plt.show
````

````python
#가로 막대그래프
#barh()메서드로 가로 막대그래프를 가짐
#오차 표시
plt.title("막대 그래프")
plt.barh(['a','b','c'],[20,78,43], xerr=[2,10,4], color = '#ff000080')
plt.show
````

### 스템 플롯
기준점을 기준으로 오차범위를 표현하고자 할때 사용하는 차트

````python
x = np.linspace(0.1 , 2 * np.pi, 10)
plt.title('스템플롯')
plt.stem(x, np.cos(x))
plt.show()
````

### 파이차트
각 분류가 전체에서 차지하는 비중을 표현 하고자 할때 사용하는 그래프

````python
labels = ['개구리', '돼지', '개', '통나무']
sizes = [15, 30, 45, 10]
colors = ['yellowgreen', 'gold', 'lightskyblue', 'lightcoral']
explode = (0, 0.1, 0, 0)
plt.title("Pie Chart")
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
        autopct='%1.1f%%', shadow=True, startangle=90)
plt.axis('equal')
plt.show()
````

### 점 그래프
변수간의 상관관계 표현할때 사용하는 그래프

````python
x = np.random.normal(0,1,100)
y1 = np.random.normal(0,1,100)
y2 = np.random.normal(0,1,100)

plt.title('점 그래프')
plt.scatter(x,y1)
plt.grid(True)
plt.show
````

````python
plt.title('버블 그래프')
plt.scatter(x,y1,c = y2)
plt.grid(True)
plt.show
````

### 컨투어 플롯
3차원 자료에서 x,y입력과 z출력이 존재 할 때, z를 선과 색상으로 표현하는 그래프

````python
def func(x,y):
    z = (1 - x / 2 + x ** 5 + y ** 3) * np.exp(-x **2 -y **2)
    return z
x = np.linspace(-3,3,256)
y = np.linspace(-3,3,256)

xx, yy = np.meshgrid(x, y)
zz = func(xx,yy)

plt.title('등고선 그래프')
plt.contour(xx,yy,zz)
plt.contourf(xx,yy,zz)
plt.grid(True)
plt.show()
````
